# 2. 금융 시계열의 특징

### 학습목차
1. 금융에서 활용되는 수익률(Return)
* 다기간 수익률(Multiperiod Simple Return)
* 다기간 로그 수익률(Multiperiod Log Return)

2. 자산의 수익률(Asset Return)에 대한 주요한 경험기반의 사실들(stylized fact)
* 경험기반의 사실(stylized fact) : 과거부터 축적 되어있는 정형화된 사실들을 기반으로 접근하는 방법론입니다.
* 좌우 대칭 분포(symmetric distribution) : 정규분포의 모양처럼 좌우가 대칭인 경우를 예시로 들 수 있습니다.
* 두꺼운 꼬리 분포(Heavy-tailed distribution) : 지수함수보다 느리게 감쇠되는 꼬리를 가지는 분포이며 정규분포에 비해 꼬리가 두꺼운 경우를 말합니다.

## 1. 금융에서 활용되는 수익률(Return)
* 금융에서는 자산의 가치 자체 보다는 수익에 대해서 더 높은 관심을 지니고 있습니다.
* 그렇기에 금융에서 활용되는 두 가지 수익률(Return)에 대해 알아볼 필요가 있습니다.    
    
아래 수식에서의 각 인자 값들은 각 시점의 가격을 의미합니다.
* 다기간 수익률(Multiperiod Simple Return)
* 다기간 로그 수익률(Multiperiod Log Return)
* 연속된 기간의 수익률을 계산할 때, 로그 수익률은 Additive하게 적용될 수 있음으로 연산이 간편합니다.
    * Additive란 다른 요인(factors)들과 결합될 때 알려진 정도보다 더 누적 효과나 강도가 좋아지는 것을 말합니다.

## 2. 자산의 수익률(Asset Return)에 대한 주요한 경험기반의 사실(stylized fact)들
* 일반적으로 평균을 중심으로 좌우로 대칭(symmetric)을 보입니다.
* 다만 굉장히 두꺼운 꼬리 분포 (Heavy-tailed distribution)를 따릅니다. (단기간에 대박나거나 쪽박나는 사람들이 있다!)
* 각 시점의 수익률들 간에는 별다른 관계가 없습니다. 그러나 수익률의 제곱 간에는 높은 관계가 존재합니다.
    * 수익률들 간에는 별다른 관계가 없다 → 효율적 시장 가설(EMH; Efficient Market Hypothesis), 주식 가격은 이용가능한 정보를 즉각적으로, 정확하게 반영된다는 것입니다.
    * 수익률의 제곱 간에는 높은 관계가 존재한다 → 조건부 이분산성(Conditional Heteroskedasticity), 자산의 수익률의 분산(=제곱)은 상수가 아니며(일정하지 않으며) 이들은 자기회귀적(Autoregressive) 특성을 지니고 있습니다.

# 3. 금융 시계열을 위한 모델, ARCH

## 1. ARCH
* 자기회귀 조건부 이분산성 모델(Auto Regressive Conditional Heteroskedasticity Model)

## 2. GARCH
* 일반화된 자기회귀 조건부 이분산성 모델(Generalized AutoRegressive Conditional Heteroskedasticity Model)
* S&P 500 returns
* IBOVESPA returns    
    

    
* 수익률의 변동성이 높아지는 시기는 하나의 군집(cluster)처럼 뭉쳐 있는 것을 볼 수 있습니다.
* 변동성에 대한 AR 모델의 적용 가능성이 있습니다.
    * ACF, PACF Plot 을 살펴볼 때, 유의미한 Lag를 확인할 수 있습니다.
